```python
#Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?
#No, in quanto per alcune varianti dello stesso Pokemon il numero di Pokedex non cambia
pokemon = pokemon.set_index("#")
```

```python
colonne = ['Name', 'Type 1', 'Type 2', 'Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']
leggendari = pokemon.loc[:,"Legendary"]
pokemon.loc[leggendari,colonne]
```

```python
erba1 = pokemon.loc[:,"Type 1"]== "Grass"
erba2 = pokemon.loc[:,"Type 2"]== "Grass"
filtro_erba = erba1 | erba2
leggendari_erba = leggendari & filtro_erba
pokemon.loc[leggendari_erba,colonne]
```

```python
fuoco1 = pokemon.loc[:,"Type 1"]== "Fire"
fuoco2 = pokemon.loc[:,"Type 2"]== "Fire"
ghiaccio1 = pokemon.loc[:,"Type 1"]== "Ice"
ghiaccio2 = pokemon.loc[:,"Type 2"]== "Ice"
filtro_fuoco = fuoco1 | fuoco2
filtro_ghiaccio = ghiaccio1 | ghiaccio2
filtro_fuoco_o_ghiaccio = filtro_fuoco | filtro_ghiaccio
leggendari_ghiaccio_o_fuoco = leggendari & filtro_fuoco_o_ghiaccio
pokemon.loc[leggendari_ghiaccio_o_fuoco,colonne]
```
